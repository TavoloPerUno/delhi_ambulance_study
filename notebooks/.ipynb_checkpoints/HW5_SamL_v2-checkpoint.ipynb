###### v. Find the solution using a Optim package, and the GoldenSection() solver option. 

```julia
using Optim, DataFrames

function f(x)
    return -x^(0.5)*(1-0.5*x)^(0.5)
end

function df(x)
    return 0.5*(x-0.5*x^2)^(-1.5)*(1-x)
end

x0=0.5
err = 1.0e8
tol = 1.0e-8

resual=DataFrame(k=NaN, x0=NaN, x=NaN, h0=NaN, err=NaN)
k=0

while err>tol
    opt=Optim.optimize(h -> f(x0+h*df(x0)), 0, 2, GoldenSection())
    h0 = opt.minimizer
    x=x1+h0*df(x0)
    err=maximum(abs.(x-x0))
    if k==0
        resual=DataFrame(k=k,x0=x0,x=x,h0=h0,err=err)
    else
        push!(resual,([k,x0,x,h0,err]))
    end
    x0=x
    k=k+1
end
resual
        
```

#### 3 - Write out the Taylor expansion of the function $f(x)=e^x$, include the first six components. Plot the function $f(x)=e^x$ and its first, second, third, forth, and fifth order approximation in the range $[0,2]$. 

$$\small{\mathrm{Taylor\ Expansion\ :}} \\ 
f(x) = f(x_0)+ e^{x_0}\epsilon + \frac{1}{2!}e^{x_0}\epsilon^2 + \frac{1}{3!}e^{x_0}\epsilon^3 + \frac{1}{4!}e^{x_0}\epsilon^4 + \frac{1}{5!}e^{x_0}\epsilon^5 + \frac{1}{6!}e^{x_0}\epsilon^6 + ...  \\
\small\mathrm{given}\quad x=x_0+\epsilon$$

```julia
using PyPlot

function approx_exp_by_order(x, n)
    if n == 0
        return 0
    else
        return x ^(n) / factorial(n) + approx_exp_by_order(x, n-1)
    end
end

x1 = collect(0:0.01:2)
f0 = e.^x1
f1 = 1+approx_exp_by_order(x1,1)
f2 = 1+approx_exp_by_order(x1,2)
f3 = 1+approx_exp_by_order(x1,3)
f4 = 1+approx_exp_by_order(x1,4)
f5 = 1+approx_exp_by_order(x1,5)
f6 = 1+approx_exp_by_order(x1,6)

fig, ax=subplots(1,1)

#ax[:set_title]("Objective Function", fontsize=7)
ax[:plot](x1, f0, color="grey",linestyle="-",linewidth=2,label=L"f(x)=e^x")
#ax[:set_title]("First Order", fontsize=7)
ax[:plot](x1, f1, color="blue",linestyle="-",linewidth=2,label=L"f(x)=1+x")
#ax[:set_title]("Second Order", fontsize=7)
ax[:plot](x1, f2, color="green",linestyle="-",linewidth=2,label=L"f(x)=1+x+\frac{1}{2!}x^2")
#ax[:set_title]("Third Order", fontsize=7)
ax[:plot](x1, f3, color="pink",linestyle="-",linewidth=2,label=L"f(x)=1+x+\frac{1}{2!}x^2+\frac{1}{3!}x^3")
#ax[:set_title]("Forth Order", fontsize=7)
ax[:plot](x1, f4, color="brown",linestyle="-",linewidth=2,label=L"f(x)=1+x+\frac{1}{2!}x^2+\frac{1}{3!}x^3+\frac{1}{4!}x^4")
#ax[:set_title]("Fifth Order", fontsize=7)
ax[:plot](x1, f5, color="purple",linestyle="-",linewidth=2,label=L"f(x)=1+x+\frac{1}{2!}x^2+\frac{1}{3!}x^3+\frac{1}{4!}x^4+\frac{1}{5!}x^5")
#ax[:set_title]("Sixth order", fontsize=7)
ax[:plot](x1, f6, color="red",linestyle="-",linewidth=2,label=L"f(x)=1+x+\frac{1}{2!}x^2+\frac{1}{3!}x^3+\frac{1}{4!}x^4+\frac{1}{5!}x^5++\frac{1}{6!}x^6")
```
